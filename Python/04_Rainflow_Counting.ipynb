```python
# Data Source 5.12 - Fundamentals of Metal Fatigue
import math
import matplotlib.pyplot as plt
raw_data = [0,-50.5,51.6,-.2,30.5,-35.8,5.3,-46.3,45.3,-20,-15.8,-22.1,49.5,-4.9,11.6,-4.2,32.6,11.6,18.9,-15.8,-8.4,-10.5,8.4,-6.3,4.2,-25.3,-20.0,-35.8]
mini = raw_data.index(min(raw_data)) # Get the index of the smallest element in the list.
sort_data = raw_data[mini:] + raw_data[0:mini] # This is done as the ASTM standard requires the cycle to start at peak or valley
plt.plot(raw_data)
plt.plot(sort_data)

i=0
disc_data =[]
lstl = len(sort_data)-3
print(lstl)
while i < lstl:
    if (abs(sort_data[i+1] - sort_data[i+2])) < (abs(sort_data[i]-sort_data[i+1])):
        i = i+1
    else:
        disc_data.append(sort_data[i+1])
        disc_data.append(sort_data[i+2])
        del sort_data[i+1]
    i = i+1
    pass
print(disc_data)
        


    
        
            
    
```
