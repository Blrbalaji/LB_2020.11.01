```python
# Object Re-usabbility Concept & Interning

# Interning: Defined as creation of objects during runtime

# Objects must be re-used if the following conditions are met
# data type is integer from -5 to 256
# data type is string "0-9" "a-z" "A-Z", _(underscore)

# data type floats, complex or derived does not follow object re-usability concept

# Example: You will notice that the memory address of both a & b are identical

a = 10
b = 10
print(id(a)) 
print(id(b))

# Example: You will notice that the memory address of both a & b are different as it exceeds 256

a = 257
b = 257
print(id(a)) 
print(id(b))

# Example: Memory address of both a & b are different as float does not follow object reusability 

a = 20.1
b = 20.1
print(id(a)) 
print(id(b))

# Example: Memory address of both a & b are same.

a = str('z')
b = str('z')
print(id(a)) 
print(id(b))
```

# # Understanding Input & Output Functions
    1. Function print()
        1. Syntax: print(object(s), sep=separator, end=end, file=file, flush=flush)
            a. sep = Optional, default ' '
            b. end = Optional, default \n line feed.
            c. file = Optional. An object with a write method. Default is sys.stdout
            d. flush = Optional. A Boolean, specifying if the output is flushed (True) or buffered (False) - Default 




```python
# Example print()

a = 10
b = 20
c = a+b

print(a)
print(b)
print(c)

# When the above statements are printed, it prints 10 20 and 30 in separate line due to default end being \n.

print(a,b,c) # statements are printed 10 20 30 as default is space.

print(a,b,c, sep = '\t') # Shows a tab separator

print(a,b,c, sep = ' @ ')

# \n is overridden by \t therefore, a, b and c are printed on same line separated by tab
print(a, end = '\t')
print(b, end = '\t')
print(c, end = '\t')

print('Sum of {} and {} = {}'.format(a,b,c)) # is same as writing
print(f'Sum of {a} and {b} = {c}')

# printing to a file: This is bit tricky you need to import sys
import sys
out = open('test.txt', 'w') # i.e. open a text by name test in write W mode. If test is NA, it will create it.
print(a,b,c, sep = '\n', file = out)
out.close()

```

```python
###  Fibonacci Sequence Using Function

def fib_seq(num):
    if num <= 1:
        return(num)
    else:
        return(fib_seq(num-1) + fib_seq(num-2))

n_terms = int(input("Enter Number of Terms in the Sequnce: "))

if n_terms <= 0:
    print('Enter a Positive Number')
else:
    for i in range(n_terms):
        print(fib_seq(i))
```

```python
### Function
def my_oper():
    num1 = float(input("Enter 1st Number"))
    num2 = float(input("Enter 2nd Number"))
    res_add = num1 + num2
    if num1 > num2:
        res_sub = num1 - num2
    else:
        res_sub = num2 - num1
    res_prod = num1 * num2
    res_trudiv = num1 / num2
    res_rema = num1 % num2
    res_quo = num1 // num2
    return res_add, res_sub, res_prod, res_trudiv, res_rema, res_quo

print(my_oper()) # Default packing is a tuple
```

```python
a = 10 # Global
b = 20 # Global

def f1():
    a = 11 # Extended
    x = 22 # Extended
    print(a, x)
    
    def f2():
        nonlocal a
        b = 33 # Local
        m = 44 # Local
        print(b, m, a)
    f2()
f1()
```

```python
"""
Version: 1.0
Created on: Sun Dec 13 21:31:49 2020
Author: Balaji Kannan
Description: Working with Core Python Libraries

"""
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from pandas_profiling import ProfileReport
from pca import pca

# File Path & Name Definitions

PATH = r"C:\DA_Case_Studies\01_Liver_Patient\Input"
FNAME = r"\indian_liver_patient.csv"

OUTPATH = r"C:\DA_Case_Studies\01_Liver_Patient\Output"
OFNAME0 = r"\00_ProfileReport.html"
OFNAME1 = r"\01_Descriptive_Stats.txt"
OFNAME2 = r"\02_Correlation_Matrix.txt"
OFNAME3 = r"\03_Variance_Inflation_Factor.txt"
OFNAME4 = r"\04_Principal_Components.txt"
FIG1 = r"\Pearson_Corr"
FIG2 = r"\Model_Plot"
FIG3 = r"\2D_BiPlot"
FIG4 = r"\3D_Biplot"

# DataFrame Definitions

df = pd.read_csv(f"{PATH}{FNAME}")
df = df.dropna()
df.columns = ['Age', 'Gender', 'TB', 'DB', 'Alkphos', 'Sgpt', 'Sgot', 'TP', 'ALB', 'AG', 'Patient']
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0})
XS = df.drop(columns=['Patient'])
ys = df.filter(['Patient'], axis=1)
XS = pd.DataFrame(data=XS, columns=['Age', 'Gender', 'TB', 'DB', 'Alkphos', 'Sgpt', 'Sgot', 'TP', 'ALB', 'AG'])
pd.options.display.max_columns = None

# Analyses

print(os.getcwd(), end = '\n')

dstat = df.describe() # Descriptive Statistics
mcorr = df.corr(method = 'pearson')

df = df[['Age', 'Gender', 'TB', 'DB', 'Alkphos', 'Sgpt', 'Sgot', 'TP', 'ALB', 'AG', 'Patient']]
vif_data = pd.DataFrame()
vif_data["Feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]

model = pca()
out = model.fit_transform(XS)
pc = out['topfeat']

# Output

report = ProfileReport(df)
report.to_file(f"{OUTPATH}{OFNAME0}")

FOUT1 = open(f"{OUTPATH}{OFNAME1}", 'w+')
dstat.to_string(FOUT1)
FOUT1.close()

FOUT2 = open(f"{OUTPATH}{OFNAME2}", 'w+')
mcorr.to_string(FOUT2)
FOUT2.close()

FOUT3 = open(f"{OUTPATH}{OFNAME3}", 'w+')
vif_data.to_string(FOUT3)
FOUT3.close()

FOUT4 = open(f"{OUTPATH}{OFNAME4}", 'w+')
pc.to_string(FOUT4)
FOUT4.close()

mask = np.zeros(mcorr.shape, dtype=bool)
mask[np.tril_indices(len(mask))] = False
plt.subplots(figsize=(20,15))
# cmap = 'coolwarm' / 'viridis'
sns.heatmap(mcorr, annot=True, vmin=-1, vmax=1, center=0, cmap='viridis', square=True, mask=mask)
plt.savefig(f"{OUTPATH}{FIG1}")

```
